###  2) 2010~2018 국내 생활 폐기물 발생현황 중 종량제방식에의한혼합배출, 재활용가능자원 분리배출, 음식물류폐기물분리배출 비율

```python
def csv_to_list(폐기물_발생현황_생활폐기물_종류별):
    import csv
    lists = []
    file = open('폐기물_발생현황_생활폐기물_종류별.csv','r')
    csvfile = csv.reader(file)
    
    for element in csvfile:
        lists.append(element)
        
    return lists

result = csv_to_list('폐기물_발생현황_생활폐기물_종류별.csv')

#print(result[0])
#print("\n")
#print(result[1])
#print("\n")
#print(result[2])
#print("\n")

#print(result[3])

total = result[3]
mix = []
recycle = []
food = []

# 리스트에서 필요한 데이터의 index를 알아냈다. ex) print(total.index('11580.5'))

i = 2
for a in range(9):
    mix.append(total[i])
    if i > 102:
        i = i + 22  
    else:
        i = i + 20
mix_float = list(map(float, mix))
print(mix_float)

j = 3
for b in range(6):
        recycle.append(total[j])
        j = j + 20    
    
recycle.append(total[124])
recycle.append(total[146])
recycle.append(total[168])
        
recycle_float = list(map(float, recycle))
print(recycle_float)

k = 20
for c in range(6):
    food.append(total[k])
    k = k + 20
    
food.append(total[142])
food.append(total[164])
food.append(total[187])

food_float = list(map(float, food))
print(food_float)
```

```python
import masspercentage  # masspercentage class를 같은 디렉토리 안에 만들어 import했다. 
#mix 종량제방식에 의한 혼합배출
ans1 = masspercentage.MassPercentage(total_float, mix_float)
result1 = ans1.percentage_of_mass()
print(result1)

#recycle 재활용가능 자원 분리배출
ans2 = masspercentage.MassPercentage(total_float, recycle_float)
result2 = ans2.percentage_of_mass()
print(result2)

#food 음식물류 폐기물 분리배출
ans3 = masspercentage.MassPercentage(total_float, food_float)
result3 = ans3.percentage_of_mass()
print(result3)
```
